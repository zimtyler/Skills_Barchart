```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Skills = ["Excel", "SQL", "Python", "R", "Tableau", "Quantitative\nModeling", "Workflow\nAutomation"]
Proficiency = [10, 9, 8, 7, 9, 10, 10]

df = pd.DataFrame(list(zip(Skills, Proficiency)), columns = ["Skills", "Proficiency"])

print(df)
```

```python
#dots per inch of 300
plt.rcParams['figure.dpi'] = 300

sns.set(style='whitegrid')

#adjust size of plot
ax = plt.subplots(figsize=(4, 6))

sns.barplot(data = df,
            x = "Proficiency", 
            y = "Skills",
            palette = 'blend:#323,#479'
            )
#Remove standard x and y axis labels and ticks
plt.xlabel("")
plt.ylabel("")
plt.xticks([], [])
plt.yticks([], [])

# loop over index to label proficiency integer within the bars, setting font to serif to match resume
for i in range(len(Proficiency)):
    plt.text(x=int(Proficiency[i])-.7, 
            y = i,
            s = Proficiency[i],
            color='white',
            fontsize=16,
            family = 'serif',
            horizontalalignment = 'center', 
            verticalalignment = 'center'
            )

#initially tried to loop, but quickly realized that I had to adjust the position of the string individually
#labeling each bar with Skill string
plt.text(x= 1.2, 
         y = 0,
         s = Skills[0],
         color = 'white',
         fontsize=15,
         family = 'serif',
         horizontalalignment = 'center', 
         verticalalignment = 'center'
         )
plt.text(x= 1, 
         y = 1,
         s = Skills[1],
         color='white',
         fontsize=15,
         family = 'serif',
         horizontalalignment = 'center', 
         verticalalignment = 'center'
            )
plt.text(x= 1.5, 
         y = 2,
         s = Skills[2],
         color='white',
         fontsize=15,
         family = 'serif',
         horizontalalignment = 'center', 
         verticalalignment = 'center'
            )
plt.text(x= .5, 
         y = 3,
         s = Skills[3],
         color='white',
         fontsize=15,
         family = 'serif',
         horizontalalignment = 'center', 
         verticalalignment = 'center'
            )
plt.text(x= 1.68, 
         y = 4,
         s = Skills[4],
         color='white',
         fontsize=15,
         family = 'serif',
         horizontalalignment = 'center', 
         verticalalignment = 'center'
            )
plt.text(x= .25, 
         y = 5,
         s = Skills[5],
         color='white',
         fontsize=15,
         family = 'serif',
         horizontalalignment = 'left', 
         verticalalignment = 'center'
            )
plt.text(x= .3, 
         y = 6,
         s = Skills[6],
         color='white',
         fontsize=15,
         family = 'serif',
         horizontalalignment = 'left', 
         verticalalignment = 'center'
            )
#positioned the "title" at the bottom, really just showing off now
plt.title("* Visualization made with Pandas, Matplotlib, & Seaborn",
          y = -.04,
          horizontalalignment = 'center',
          fontsize = 7,
          family = 'serif'
          )

#Removed the x and y axis for a polished looked
sns.despine(left=True,
            bottom = True
            )
```
